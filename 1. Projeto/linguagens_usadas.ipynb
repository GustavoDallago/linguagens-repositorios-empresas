o objetivo do projeto é para encontrar quais são as linguagens de programação utilizadas por algumas grandes empresas.

# Extraindo os dados

### Obtendo dados dos repositórios

```python
# Importanto biblioteca
import requests
```

```python
# Especificando a versão da API
headers = {'X-Github-Api-Version': '2022-11-28'}
```

```python
api_base_url = 'https://api.github.com'
owner ='amzn' #username da amazon
url = f'{api_base_url}/users/{owner}/repos'
```

```python
response = requests.get(url, headers=headers)
response.status_code
```

### Autênticação

```python
access_token = 'Insira um token'
headers = {'Authorization': f'Bearer '+ access_token,
           'X-Github-Api-Version': '2022-11-28'}
```

### Paginando os repositórios

```python
repos_list = []

for page_num in range (1,6):
    try:
        url_page =f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

# Transformando os dados

### Nome dos repositórios

```python
repos_name = []

for repo in repos_list:
    for item in repo:
        repos_name.append(item['name'])
```

### Linguagem dos repositórios

```python
repos_language = []

for repo in repos_list:
    for item in repo:
        repos_language.append(item['language'])
```

### Criando um DataFrame

```python
import pandas as pd
```

```python
dados_amz = pd.DataFrame()
dados_amz['repository_name'] = repos_name
dados_amz['repository_language'] = repos_language
dados_amz.head()
```

### Salvando o DataFrame

```python
dados_amz.to_csv('2. Output_final/amazon.csv', index=False)
```

# Armazenando os dados

### Criando repositório com POST

```python
api_base_url = 'https://api.github.com'
url = f'{api_base_url}/user/repos'
```

```python
data = {
    'name': 'Linguagens-utilizadas',
    'description': 'Repositório para armazenar os dados de linguagens utilizadas nos repositórios da Amazon',
    'private': False
}

response = requests.post(url, json=data, headers=headers)
response.status_code
```

### Formato do arquivo

```python
import base64   
```

```python
with open('2. Output_final/amazon.csv', 'rb') as file:
    file_content = file.read()
    encoded_content = base64.b64encode(file_content)
```

### Upload de arquivo com PUT

```python
api_base_url = 'https://api.github.com'
username = 'GustavoDallago'
repo_name = 'Linguagens-utilizadas'
path = '2. Output_final/amazon.csv'

url = f'{api_base_url}/repos/{username}/{repo_name}/contents/{path}'
url
```

```python
data = {
    'message': 'Adicionando arquivo amazon.csv',
    'content': encoded_content.decode('utf-8')
}

response = requests.put(url, json=data, headers=headers)
response.status_code
```
